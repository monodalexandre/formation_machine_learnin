## 1 - Data cleaning

```python
# Check data

print(len(list(raw.columns)))
pd.set_option('display.max_columns', 200)
raw.head()
```

Too many columns. Need to drop some.

```python
raw.describe()
```

```python
# Drop first technical columns
raw = raw.drop(columns=raw.columns[0:7])

# Drop empty/almost-empty columns
threshold_columns = raw.count().max()/100*10    # Drop columns with 10% values or less
raw = raw.drop(columns=raw.columns[raw.count()<=threshold_columns])

# Drop tags columns
raw = raw[[x for x in raw if not x.endswith("tags")]]

# Drop columns with images
raw = raw[[x for x in raw if not x.startswith("image")]]

# When duplicates, drop columns that have non fr name
raw = raw.drop(columns=["categories", 
        "labels", "countries", 'additives', "states","main_category", "nutrition-score-uk_100g"])
```
